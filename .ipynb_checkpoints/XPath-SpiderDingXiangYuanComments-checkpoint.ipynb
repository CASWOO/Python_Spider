# BeautifulSoup——中国大学排名

```python
import requests
from bs4 import BeautifulSoup
import bs4
r = requests.get('http://www.zuihaodaxue.cn/zuihaodaxuepaiming2019.html')
r.encoding="utf8" #让beautifulsoup提取出的页面中的中文字符不乱码
demo = r.text
soup = BeautifulSoup(demo,'html.parser')
soup
```

```python
len(soup.find_all("tr","alt"))
```

```python
array = [ [ 0 for i in range(3)] for i in range(len(soup.find_all("tr","alt")))]
```

```python
index=0
for node in soup.find_all("tr","alt"):
    array[index][0]=node.find_all('td')[0].string
    array[index][1]=node.find_all('td')[1].string
    array[index][2]=node.find_all('td')[3].string
    index=index+1
array
```

```python
import pandas as pd
c={"排名" :  [i[0] for i in array] ,
   "学校" :  [i[1] for i in array],
   "总分" :  [i[2] for i in array]}#将列表a，b转换成字典
pd.DataFrame(c)
```
